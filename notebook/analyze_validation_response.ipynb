```python
import pandas as pd

df = pd.read_csv('resopnse_validation_parsed_20240226_1614.csv')
```

```python
grouped = df.groupby('rowid')

aggregated_cause1_icd10_rows = []
aggregated_cause2_icd10_rows = []
aggregated_cause3_icd10_rows = []
aggregated_cause4_icd10_rows = []
aggregated_cause5_icd10_rows = []

for name, group in grouped:
    # print(f"Row ID: {name}", group['cause1_icd10'].value_counts().to_dict())
    aggregated_cause1_icd10_rows.append([name, group['cause1_icd10'].value_counts().to_dict()])
    aggregated_cause2_icd10_rows.append([name, group['cause2_icd10'].value_counts().to_dict()])
    aggregated_cause3_icd10_rows.append([name, group['cause3_icd10'].value_counts().to_dict()])
    aggregated_cause4_icd10_rows.append([name, group['cause4_icd10'].value_counts().to_dict()])
    aggregated_cause5_icd10_rows.append([name, group['cause5_icd10'].value_counts().to_dict()])
    
df1 = pd.DataFrame(aggregated_cause1_icd10_rows, columns=['rowid', 'cause1_icd10']).set_index('rowid')
df2 = pd.DataFrame(aggregated_cause2_icd10_rows, columns=['rowid', 'cause2_icd10']).set_index('rowid')
df3 = pd.DataFrame(aggregated_cause3_icd10_rows, columns=['rowid', 'cause3_icd10']).set_index('rowid')
df4 = pd.DataFrame(aggregated_cause4_icd10_rows, columns=['rowid', 'cause4_icd10']).set_index('rowid')
df5 = pd.DataFrame(aggregated_cause5_icd10_rows, columns=['rowid', 'cause5_icd10']).set_index('rowid')

combined_df = df1.join([df2, df3, df4, df5])

pd.set_option('display.max_rows', None)

combined_df
```
